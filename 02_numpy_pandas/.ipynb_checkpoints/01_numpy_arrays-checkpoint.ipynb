# AI Engineer Numpy Arrays

```python
# creating arrays
import numpy as np
onedarr=np.arange(1,21)
matrix=np.zeros((3,3))
eye=np.eye(3)
evenly_space=np.linspace(0,1,10)
print(f"1D Array — shape: {onedarr.shape}, dtype: {onedarr.dtype}, size: {onedarr.size}")
print(f"Zeros — shape: {matrix.shape}, dtype: {matrix.dtype}, size: {matrix.size}")
print(f"Identity — shape: {eye.shape}, dtype: {eye.dtype}, size: {eye.size}")
print(f"Linspace — shape: {evenly_space.shape}, dtype: {evenly_space.dtype}, size: {evenly_space.size}")
```

```python
# slicing 
scores = np.array([0.95, 0.82, 0.74, 0.61, 0.50, 0.43, 0.38, 0.91, 0.88, 0.76])
print(scores[0:3])
print(scores[-3:])   #  always last 3  
print(scores[0::2])
print(scores[::-1])
print(scores[3:8])
```

```python
# statistical operations
scores = np.array([0.95, 0.82, 0.74, 0.61, 0.50, 0.43, 0.38, 0.91, 0.88, 0.76])
print(scores.mean())
print(scores.std())
print(scores.max())
print(scores.min())
print(scores.sum())
print(f"Best score index: {scores.argmax()} — value: {scores[scores.argmax()]}")
print(f"Worst score index: {scores.argmin()} — value: {scores[scores.argmin()]}")
```

```python
# reshape
import numpy as np

# Create 2D array (4x3)
arr = np.arange(12).reshape(4, 3)
print("Full array:")
print(arr)

# Entire second row
print(arr[1])

# Entire third column
print(arr[:, 2])

# Top-right 2x2 submatrix
print(arr[:2, 1:])
```

```python
# boolean indexing
scores = np.array([0.55, 0.82, 0.90, 0.48, 0.76, 0.95, 0.61, 0.88])
print(scores[scores>0.75])
print(scores[(scores>0.70)&(scores<0.90)])
scores[scores < 0.60] = 0.0

print(scores)
```
