```python
import numpy as np
matrix=np.random.rand(3,4)
print(matrix)
new_matrix=matrix+10
print(new_matrix)
col_means=np.mean(new_matrix,axis=0)
print(col_means)
new_col_means=new_matrix-col_means
print(new_col_means)
col_std=np.std(new_matrix,axis=0)
print("Before normalization:")
print(matrix)
mean = matrix.mean(axis=0)
std = matrix.std(axis=0)
normalized = (matrix - mean) / std
print("After normalization:")
print(normalized)
```

```python
# task 2
data = np.array([[85, 90, 78],
                 [92, 88, 95],
                 [70, 75, 80],
                 [88, 92, 85]])
avg_score_col=np.mean(data,axis=1)
avg_score_mean=np.mean(data,axis=0)
top_student=np.argmax(avg_score_col)
hardest_subject = np.argmin(avg_score_mean)  # ✅ subject with lowest mean
print(f"Hardest subject index: {hardest_subject}")
print(avg_score_col)
print(avg_score_mean)
print(top_student)
```

```python
# transpose and inverse
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
c=A@B
A_T=A.T
A_inv = np.linalg.inv(A)
I = A @ A_inv
print("A @ B =\n", c)
print("Transpose of A =\n", A_T)
print("Inverse of A =\n", A_inv)
print("A @ A_inv (should be Identity) =\n", I)
```

```python
# functions without libraries
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def mse_loss(predictions, targets):
    return np.mean((predictions - targets) ** 2)
def normalize(arr):
    mean = np.mean(arr)
    std = np.std(arr)
    return (arr - mean) / std

# Inputs (like logits in ML)
x = np.array([-2.0, 0.0, 2.0, 4.0])

# Predictions vs Targets
predictions = np.array([0.1, 0.4, 0.8, 0.9])
targets = np.array([0, 0, 1, 1])

# Feature values
features = np.array([10, 20, 30, 40, 50])

# Apply functions
print("Sigmoid:", sigmoid(x))
print("MSE Loss:", mse_loss(predictions, targets))
print("Normalized:", normalize(features))
```

```python
# loop vs vectorization
import numpy as np
import time

# Step 1: Create array of 1 million random numbers
arr = np.random.rand(1_000_000)

# -----------------------------
#  Python Loop
# -----------------------------
start = time.time()

squared_loop = []
for x in arr:
    squared_loop.append(x ** 2)

end = time.time()
loop_time = end - start

print("Loop Time:", loop_time)


# -----------------------------
#  NumPy Vectorized
# -----------------------------
start = time.time()

squared_vectorized = arr ** 2

end = time.time()
vector_time = end - start

print("Vectorized Time:", vector_time)


# -----------------------------
#  Speed Comparison
# -----------------------------
speedup = loop_time / vector_time

print("NumPy is faster by:", speedup, "times")
```
