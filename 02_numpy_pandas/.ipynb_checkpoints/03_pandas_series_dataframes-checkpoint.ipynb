# AI Engineer Pandas series and dataframe

```python
import pandas as pd

data = {
    "model": [
        "Logistic Regression",
        "Random Forest",
        "XGBoost",
        "ResNet50",
        "BERT",
        "GPT-3"
    ],
    "accuracy": [
        0.85,
        0.89,
        0.91,
        0.93,
        0.94,
        0.96
    ],
    "training_time_seconds": [
        5,
        120,
        180,
        3600,
        7200,
        200000
    ],
    "framework": [
        "scikit-learn",
        "scikit-learn",
        "XGBoost",
        "PyTorch",
        "TensorFlow",
        "OpenAI"
    ],
    "year_released": [
        1958,
        2001,
        2016,
        2015,
        2018,
        2020
    ]
}

df = pd.DataFrame(data)
print("Shape:", df.shape)
print("\nDtypes:\n", df.dtypes)
print("\nHead:\n", df.head())
print("\nDescribe:\n", df.describe())
```

```python
# task 2 
df_selected = df[["model", "accuracy"]]
print(df_selected)
df_filter=df[df["accuracy"]>0.90]
print(df_filter)
filter_framework=df[df["framework"]=="PyTorch"]
print(filter_framework)
top_3 = df.sort_values(by="accuracy", ascending=False).head(3)
print(top_3)
```

```python
# adding new cols to dataframe
import numpy as np
df["accuracy_pct"]=df["accuracy"]*100
df["grade"] = np.where(
    df["accuracy"] >= 0.92, "A",
    np.where(df["accuracy"] >= 0.85, "B", "C")
)
df["is_modern"] = df["year_released"] >= 2020
df
```

```python
# get info
value_framework=df["framework"].value_counts() 
print(value_framework)
value_grade = df["grade"].value_counts() 
print(value_grade)
unique_framework=np.unique(df["framework"])
print(unique_framework)
df_sorted = df.sort_values(by="training_time_seconds", ascending=True)
print(df_sorted)
```

```python
# saving and reloading
df.to_csv("models.csv", index=False)
import pandas as pd

df2 = pd.read_csv("models.csv")
df2.info()
print(df.equals(df2))
```
