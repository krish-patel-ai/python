# AI Engineer Groupby

```python
# groupby
import pandas as pd
data = {
    "model":     ["BERT", "XGBoost", "ResNet", "GPT-3", "LinearReg", "XGBoost", "BERT", "ResNet"],
    "framework": ["PyTorch", "XGBoost", "PyTorch", "OpenAI", "Sklearn", "XGBoost", "PyTorch", "PyTorch"],
    "task":      ["NLP", "tabular", "CV", "NLP", "tabular", "tabular", "NLP", "CV"],
    "accuracy":  [0.94, 0.91, 0.93, 0.96, 0.78, 0.89, 0.92, 0.90],
    "epochs":    [10, 100, 50, 30, 200, 80, 15, 45]
}
df=pd.DataFrame(data)
mean_accuracy=df.groupby("framework")["accuracy"].mean()
max_accuracy=df.groupby("framework")["accuracy"].max()
total_epochs=df.groupby("framework")["epochs"].sum()
model_count=df.groupby("framework")["model"].count()
print("Mean accuracy per framework:\n", mean_accuracy)
print("\nMax accuracy per framework:\n", max_accuracy)
print("\nTotal epochs per framework:\n", total_epochs)
print("\nModel count per framework:\n", model_count)
```

```python
# using the agg function
result = df.groupby("framework")["accuracy"].agg(["mean", "max", "min",])
print(result)
```

```python
result = df.groupby(["framework","task"])["accuracy"].mean()
print(result)
```

```python
df["mean_acc_by_framework"] = df.groupby("framework")["accuracy"].transform("mean")
print(df)
df_filtered = df.groupby("framework").filter(lambda x: len(x) > 1)
print(df_filtered)
```

```python
# merging the data
experiments = pd.DataFrame({
    "exp_id":    [1, 2, 3, 4, 5],
    "model":     ["BERT", "XGBoost", "ResNet", "GPT-3", "LinearReg"],
    "framework": ["PyTorch", "XGBoost", "PyTorch", "OpenAI", "Sklearn"]
})

metrics = pd.DataFrame({
    "exp_id":   [1, 2, 3, 4, 6],
    "accuracy": [0.94, 0.91, 0.93, 0.96, 0.85],
    "loss":     [0.21, 0.34, 0.28, 0.15, 0.42]
})
merged=pd.merge(experiments,metrics)
inner_join=pd.merge(experiments,metrics,how="inner")
left_join=pd.merge(experiments,metrics,how="left")
print(merged)
print(inner_join)
print(left_join)

# Inner Join:
# Only keeps rows where exp_id exists in BOTH DataFrames.
# exp_id = 6 is dropped (not in experiments)
# exp_id = 5 is dropped (not in metrics)

# Left Join:
# Keeps ALL rows from experiments (left table).
# If matching exp_id not found in metrics, fills with NaN.
# So exp_id = 5 remains, but accuracy and loss become NaN.s
```
