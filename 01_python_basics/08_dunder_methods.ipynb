```python
# task 1 using str and repr
class MLModel:
    def __init__(self,name,accuracy):
        self.name=name
        self.accuracy=accuracy
    def __str__(self):
        return(f"MLModel : {self.name}")
    def __repr__(self):
        return f"MLModel('{self.name}',{self.accuracy})"
model=MLModel("BERT",0.94)
print(model)
print(repr(model))
```

```python
# task 2 using len and getitem
class DataSet:
    def __init__(self,name,samples):
        self.name=name
        self.samples=samples
    def __len__(self):
        return len(self.samples)
    def __getitem__(self,index):
        return self.samples[index]
data=DataSet("TrainData",[10,20,30,40,50])
print(len(data))
print(data[2])
```

```python
# task 3 using eq,lt,gt
class ModelScore:
    def __init__(self,name,score):
        self.name=name
        self.score=score
    def __eq__(self, other):       # == operator
        return self.score == other.score

    def __lt__(self, other):       # < operator
        return self.score < other.score

    def __gt__(self, other):       # > operator
        return self.score > other.score
m1 = ModelScore("BERT", 0.94)
m2 = ModelScore("XGBoost", 0.91)
print(m1 == m2)   # False
print(m1 > m2)    # True
print(m1 < m2)
```

```python
# task 4 using add and contains
class PipeLine:
    def __init__(self,steps):
        self.steps=steps
    def __add__(self,other):
        return PipeLine(self.steps+other.steps)
    def __len__(self):
        return len(self.steps)
    def __contains__(self,item):
        return item in self.steps
p1=PipeLine(["load","clean"])
p2=PipeLine(["train","evaluate"])
p3=p1+p2
print(p3.steps)
print(len(p3))
print("train" in p3)
```
