# AI Engineer python Control flow

```python
#Task1
loss = 0.45
if loss > 0.8:
    print("Critical")
elif loss > 0.5:
    print("High")
elif loss > 0.2:
    print("Moderate")
else:
    print("Low")
```

```python
# task 2 for loop
epochs=[1,2,3,4,5]
for _,value in enumerate(epochs):
    print(f"epochs : {value}/{len(epochs)}")
```

```python
# task 3 while loop
loss=1.0
while loss > 0.05:
    loss *= 0.6
    print(f"Loss: {loss:.4f}")
```

```python
# task 4
scores = [0.55, 0.82, 0.90, 0.48, 0.76, 0.95]
newscores=[score for score in scores if score > 0.75]
print(newscores)
```

```python
# task 5
for updatedscore in scores:
    if updatedscore < 0.70:
        continue
    if updatedscore >= 0.93:
        break
    print(f"Good score: {updatedscore}")    
```
