# AI Engineer Python OOP classes

```python
# task1
class Student:
    def __init__(self,name,age,course):
        self.name=name
        self.age=age
        self.course=course

    def introduce(self):
        print(f"Hi I am {self.name} age {self.age} , studying {self.course}")
obj1=Student("Krish","21","AI Engineering")
obj2=Student("Krishp","21","AI Engineering with python")
obj1.introduce()
obj2.introduce()
```

```python
# task 2
class MLModel:
    def __init__(self,name,accuracy,epochs):
        self.name=name
        self.accuracy=accuracy
        self.epochs=epochs
    def summary(self):
        print(f" Name : {self.name}")
        print(f" Accuracy : {self.accuracy*100:.1f}")
        print(f" Epochs : {self.epochs}")
    def is_ready(self):
        return self.accuracy>=0.90
    def __str__(self):
        return f"MLModel : {self.name} ({self.accuracy*100:.1f}%)"
obj1=MLModel("linearRegression",0.92,10)
obj2=MLModel("logisticRegression",0.80,50)
obj3=MLModel("Xgboost",0.94,100)
obj1.summary()
obj2.summary()
obj3.summary()
print(obj1.is_ready())
print(obj2.is_ready())
print(obj3.is_ready())
print(obj1)
print(obj2)
print(obj3)
```

```python
# task3 
class Dataset:
    format="CSV"
    def __init__(self,name,size,features):
        self.name=name
        self.size=size
        self.features=features
    def describe(self):
        print(f" Class attribute : {self.format}")
        print(f" Name : {self.name}")
        print(f" size : {self.size}")
        print(f" features : {self.features}")
    def is_large(self):
        return self.size>10000
obj1=Dataset("Titanic",20000,12)
obj2=Dataset("Diabetes",8000,12)
obj1.describe()
print(obj1.is_large())
obj2.describe()
print(obj2.is_large())
print(obj1.format)
print(obj2.format)
```

```python
# task 4
class TrainingRun:
    def __init__(self,run_id,model_name,train_acc,val_acc):
        self.run_id=run_id
        self.model_name=model_name
        self.train_acc=train_acc
        self.val_acc=val_acc
    def is_overfitting(self):
        return (self.train_acc-self.val_acc)>0.10
    def __str__(self):
        return f"Run {self.run_id} : {self.model_name}"
obj1=TrainingRun("run_001","Xgboost",0.95,0.80)
obj2=TrainingRun("run_002","randomforest",0.88,0.85)
print(obj1)
print(obj2)
print(obj1.is_overfitting())
print(obj2.is_overfitting())

```
