# AI Engineer Python Lambda Map filter

```python
# task 1 using lambda function
categorize = lambda acc: "High" if acc >= 0.90 else ("Medium" if acc >= 0.75 else "Low")

# Testing
print(categorize(0.92))  # High
print(categorize(0.80))  # Medium
print(categorize(0.60))  # Low
```

```python
# task 2 using map()
losses = [0.95, 0.82, 0.74, 0.61, 0.50, 0.43, 0.38]
newloss=list(map(lambda s:round(s*100,1),losses))
print(newloss)
```

```python
# task 3 using filter
scores = [0.55, 0.82, 0.90, 0.48, 0.76, 0.95, 0.61, 0.88]
newscores=list(filter(lambda a: a>0.8,scores))
print(newscores)
scores_mapped = list(map(lambda b: round(b * 100, 1), newscores))
print(scores_mapped)
```

```python
# task 4 sorted
models = [
    {"name": "XGBoost", "accuracy": 0.94, "speed": 15},
    {"name": "BERT", "accuracy": 0.91, "speed": 120},
    {"name": "LinearReg", "accuracy": 0.78, "speed": 2},
    {"name": "RandomForest", "accuracy": 0.88, "speed": 30},
]
sorted_accuracy=sorted(models,key=lambda c:c['accuracy'],reverse=True)
for m in sorted_accuracy:
    print(f"name : {m['name']} and accuracy : {m['accuracy']}")
sorted_speed = sorted(models, key=lambda m: m["speed"])
for k in sorted_speed:
    print(f"name : {k['name']} and speed : {k['speed']}")
```

```python
# task 5 zip()
models  = ["BERT", "XGBoost", "SVM", "LinearReg"]
acc     = [0.94, 0.91, 0.85, 0.78]
trained = [True, True, False, True]
for model, accuracy, train in zip(models, acc, trained):
    if train:
        print(f"model : {model} | acc : {accuracy} | trained : {train}")
combined=list(zip(models,acc,trained))
print(combined)
```

```python
model_back,accuracy_back,train_back=zip(*combined)
print(model_back)
print(accuracy_back)
print(train_back)
```
