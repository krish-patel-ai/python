```python
#task 1
def save_results(filename,results):
    with open(filename,"w") as f:
        for result in results:
            f.write(result + '\n')
    with open(filename,'r') as f:
        for line in f:
            print(line.strip())
results=["Model:XGBoost","Accuracy:0.94","Loss:0.21","Epochs:100"]
save_results("experiment_log.txt",results)
```

```python
# task 2 dict
import json
def save_config(config, filename):
    with open(filename, "w") as f:
        json.dump(config, f, indent=4)
def load_config(filename):
    with open(filename, "r") as f:
        return json.load(f)
model_config = {"model": "BERT", "lr": 0.0001, "epochs": 50, "batch_size": 32}
save_config(model_config, "config.json")

loaded_config = load_config("config.json")

print(loaded_config)
```

```python
# task 3
def safe_divide(a, b):
    try:
        return a/b
    except ZeroDivisionError:
        return None
print(safe_divide(10, 2))  # Normal case
print(safe_divide(10, 0))  # Division by zero
        
```

```python
def validate_lr(lr):
    if lr <= 0 or lr > 1:
        raise ValueError("Learning rate must be between 0 and 1 (exclusive of 0).")
    return True


# Test cases
try:
    validate_lr(-0.1)
except ValueError as e:
    print("Error:", e)

try:
    validate_lr(0.001)
    print("Valid learning rate")
except ValueError as e:
    print("Error:", e)
```
