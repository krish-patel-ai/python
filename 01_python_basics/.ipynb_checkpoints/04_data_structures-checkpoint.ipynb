```python
# task 3 dictionary
experiment = {"model": "XGBoost", "lr": 0.01, "epochs": 100, "accuracy": 0.91}
experiment["epochs"]=200
experiment["batch_size"]=32
del experiment["lr"]
for key,value in experiment.items():
    print(f"Key : {key} and value is : {value}")
```

```python
# task 4 
predictions = ["spam", "ham", "spam", "spam", "ham", "phishing", "ham"]
uniques=set(predictions)
print(uniques)
train_labels = {"spam", "ham"}
all_labels = {"spam", "ham", "phishing"}
new_labels=all_labels-train_labels
print(new_labels)
```

```python
# task5 zip func
models = ["LinearReg", "RandomForest", "XGBoost", "SVM"]
scores = [0.78, 0.91, 0.94, 0.85]
for model, score in zip(models,scores):
    print(f"Model : {model} and scores : {score}")
model_scores={model: score for model,score in zip(models,scores)}
filtered={model:score for model,score in model_scores.items() if score>0.88}
print(filtered)
```

```python
# task 6
all_scores = [0.55, 0.82, 0.90, 0.48, 0.76, 0.95, 0.61, 0.88]
filteredone={round(score,1) for score in all_scores if score > 0.75}
print(filteredone)
```
