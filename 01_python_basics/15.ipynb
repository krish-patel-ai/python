```python
# task 4
import pandas as pd

# Create DataFrame
data = {
    "model": ["ResNet50", "BERT", "GPT-2", "EfficientNet"],
    "accuracy": [0.92, 0.94, 0.89, 0.93],
    "parameters": ["25M", "110M", "117M", "5M"],
    "framework": ["PyTorch", "TensorFlow", "PyTorch", "TensorFlow"]
}

df = pd.DataFrame(data)

# Print full DataFrame
print("Full DataFrame:\n")
print(df)

# Filter models with accuracy > 0.90
filtered_df = df[df["accuracy"] > 0.90]

print("\nFiltered (accuracy > 0.90):\n")
print(filtered_df)
```

## My Understanding of Key Libraries

**NumPy** — a numerical library used for working with arrays and performing mathematical operations like mean, matrix multiply and random number generation.

**Pandas** — a library for working with tabular data like CSV and Excel files. Used to load, clean and explore messy datasets.

**PyTorch** — a deep learning framework used to build neural networks. It uses tensors similar to NumPy arrays but can run on GPU for fast training.
