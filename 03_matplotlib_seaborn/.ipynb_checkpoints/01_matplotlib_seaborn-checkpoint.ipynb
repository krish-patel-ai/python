### AI Engineer Matplotlib and seaborn

```python
# line plot
import matplotlib.pyplot as plt
import seaborn as sns
epochs     = list(range(1, 11))
train_loss = [0.95, 0.82, 0.71, 0.62, 0.54, 0.47, 0.41, 0.36, 0.32, 0.28]
val_loss   = [0.98, 0.87, 0.76, 0.68, 0.61, 0.55, 0.51, 0.48, 0.46, 0.45]
plt.figure(figsize=(10,5))
plt.plot(epochs, train_loss, marker='o', linewidth=2, color='blue', label='Training Loss')
plt.plot(epochs, val_loss, marker='s', linewidth=2, linestyle='--', color='orange', label='Validation Loss')
plt.title("Training vs Validation Loss Over Epochs", fontsize=14)
plt.xlabel("Epochs", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# bar plot 
import matplotlib.pyplot as plt

# Data
models = ["BERT", "XGBoost", "ResNet", "GPT-3", "LinearReg"]
accuracies = [0.94, 0.91, 0.93, 0.96, 0.78]

# Sort models by accuracy (highest → lowest)
sorted_data = sorted(zip(models, accuracies), key=lambda x: x[1], reverse=True)
models_sorted, acc_sorted = zip(*sorted_data)

# Create figure
plt.figure(figsize=(8, 5))

# Bar chart
bars = plt.bar(models_sorted, acc_sorted)

# Title and labels
plt.title("Model Accuracy Comparison", fontsize=14)
plt.xlabel("Models", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)

# Add value labels on top
for bar, acc in zip(bars, acc_sorted):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height(),
             f"{acc:.2f}",
             ha='center', va='bottom')

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# sub plots
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- Data Setup ---
# 1. Mock epoch data for line plots and scatter plot
epochs = np.arange(1, 11)
train_loss = np.array([1.50, 1.10, 0.85, 0.65, 0.50, 0.40, 0.32, 0.25, 0.20, 0.15])
train_acc = np.array([0.55, 0.68, 0.75, 0.82, 0.86, 0.89, 0.92, 0.94, 0.95, 0.97])

# 2. Model data for the bar chart 
models = ["BERT", "XGBoost", "ResNet", "GPT-3", "LinearReg"]
accuracies = [0.94, 0.91, 0.93, 0.96, 0.78]

# Sort models highest to lowest
df = pd.DataFrame({'Model': models, 'Accuracy': accuracies})
df = df.sort_values(by='Accuracy', ascending=False)

# --- Create 2x2 Subplots ---
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Top left: Training loss line plot
axs[0, 0].plot(epochs, train_loss, marker='o', color='crimson', linewidth=2)
axs[0, 0].set_title('Training Loss Over Epochs', fontsize=14)
axs[0, 0].set_xlabel('Epoch', fontsize=12)
axs[0, 0].set_ylabel('Loss', fontsize=12)
axs[0, 0].grid(True, linestyle='--', alpha=0.6)

# Top right: Training accuracy line plot
axs[0, 1].plot(epochs, train_acc, marker='s', color='seagreen', linewidth=2)
axs[0, 1].set_title('Training Accuracy Over Epochs', fontsize=14)
axs[0, 1].set_xlabel('Epoch', fontsize=12)
axs[0, 1].set_ylabel('Accuracy', fontsize=12)
axs[0, 1].grid(True, linestyle='--', alpha=0.6)

# Bottom left: Bar chart of model accuracies
bars = axs[1, 0].bar(df['Model'], df['Accuracy'], color='skyblue', edgecolor='black')
axs[1, 0].set_title('Model Accuracies Comparison', fontsize=14)
axs[1, 0].set_xlabel('Model', fontsize=12)
axs[1, 0].set_ylabel('Accuracy', fontsize=12)
axs[1, 0].set_ylim(0, 1.1)
axs[1, 0].grid(axis='y', linestyle='--', alpha=0.6)

# Add value labels to the bar chart
for bar in bars:
    yval = bar.get_height()
    axs[1, 0].text(bar.get_x() + bar.get_width()/2, yval + 0.02, f"{yval:.2f}", 
                   ha='center', va='bottom', fontweight='bold')

# Bottom right: Scatter plot of loss vs accuracy
axs[1, 1].scatter(train_loss, train_acc, color='darkorchid', s=80, edgecolor='black', alpha=0.8)
axs[1, 1].set_title('Loss vs. Accuracy', fontsize=14)
axs[1, 1].set_xlabel('Training Loss', fontsize=12)
axs[1, 1].set_ylabel('Training Accuracy', fontsize=12)
axs[1, 1].grid(True, linestyle='--', alpha=0.6)

# Formatting
plt.tight_layout(pad=3.0)

# Display or save the plot
plt.show()[0,0].show()
```

```python
# heat map
import pandas as pd
df = pd.DataFrame({
    "model":     ["BERT", "XGBoost", "ResNet", "GPT-3", "LinearReg", "XGBoost", "BERT", "ResNet"],
    "framework": ["PyTorch", "XGBoost", "PyTorch", "OpenAI", "Sklearn", "XGBoost", "PyTorch", "PyTorch"],
    "accuracy":  [0.94, 0.91, 0.93, 0.96, 0.78, 0.89, 0.92, 0.90],
    "loss":      [0.21, 0.34, 0.28, 0.15, 0.65, 0.38, 0.24, 0.30]
})
plt.figure(figsize=(6, 4))
sns.barplot(data=df,x="framework",y="accuracy",color="blue")
plt.title("Model Accuracy Comparison")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="framework", y="accuracy")
plt.title("Accuracy Distribution by Framework")
plt.show()

plt.figure(figsize=(6, 4))
corr = df[["accuracy", "loss"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# side by side accuracy and loss using functions
import matplotlib.pyplot as plt

def plot_training_curves(train_loss, val_loss, train_acc, val_acc):
    """
    Plots training and validation loss and accuracy curves side by side.
    """
    epochs = range(1, len(train_loss) + 1)
    
    # Create a 1x2 subplot grid
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    
    # --- Left Plot: Loss ---
    axs[0].plot(epochs, train_loss, label='Training Loss', marker='o', linewidth=2)
    axs[0].plot(epochs, val_loss, label='Validation Loss', marker='s', linewidth=2)
    axs[0].set_title('Loss vs. Epochs', fontsize=14)
    axs[0].set_xlabel('Epoch', fontsize=12)
    axs[0].set_ylabel('Loss', fontsize=12)
    axs[0].legend(fontsize=10)
    axs[0].grid(True, linestyle='--', alpha=0.6)
    
    # --- Right Plot: Accuracy ---
    axs[1].plot(epochs, train_acc, label='Training Accuracy', marker='o', linewidth=2)
    axs[1].plot(epochs, val_acc, label='Validation Accuracy', marker='s', linewidth=2)
    axs[1].set_title('Accuracy vs. Epochs', fontsize=14)
    axs[1].set_xlabel('Epoch', fontsize=12)
    axs[1].set_ylabel('Accuracy', fontsize=12)
    axs[1].legend(fontsize=10)
    axs[1].grid(True, linestyle='--', alpha=0.6)
    
    # Formatting
    plt.tight_layout()
    plt.show()

# --- Generate Realistic Data for 10 Epochs ---
# Training improves steadily
t_loss = [1.20, 0.90, 0.70, 0.55, 0.45, 0.38, 0.32, 0.28, 0.25, 0.22]
t_acc  = [0.60, 0.72, 0.80, 0.85, 0.88, 0.90, 0.92, 0.93, 0.94, 0.95]

# Validation plateaus and slightly worsens at the end (mimicking overfitting)
v_loss = [1.15, 0.88, 0.72, 0.60, 0.55, 0.53, 0.55, 0.58, 0.62, 0.65]
v_acc  = [0.62, 0.74, 0.79, 0.83, 0.85, 0.86, 0.86, 0.85, 0.84, 0.84]

# Call the function
plot_training_curves(t_loss, v_loss, t_acc, v_acc)
```
