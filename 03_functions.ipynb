```python
# task1 
def describe_model(model_name,accuracy,epochs):
    print(f"Model : {model_name} | accuracy : {accuracy*100:.1f}% | trained for {epochs} epochs")

describe_model("LinearReg",0.92,100)
```

```python
# task2 
def calculate_loss(method="mse"):
    if method=="mse":
        print("Calculating MSE loss")
    elif method=="cross_entropy":
        print("Calculating cross_entropy loss")
    else:
        print("Unknown loss method")

calculate_loss("mse")
calculate_loss("cross_entropy")
calculate_loss("mae")
    
```

```python
# task3 
def train_summary(**kwargs):
    for key,value in kwargs.items():
        print(f"Key : {key} | value : {value}")
train_summary(lr=0.001, epochs=50, batch_size=32, optimizer="adam")
```

```python
# task4 
def evaluate_model(scores):
    best_score=max(scores)
    avg_score=sum(scores)/len(scores)
    return best_score,avg_score
scores=[0.85,0.90,0.70,0.92]
best,avg=evaluate_model(scores)
print(f"Best Score : {best}")
print(f"average is : {avg}")
```

```python
# task 5
def is_overfitting(train_acc, val_acc):
    """
    Determines whether a model is overfitting.

    Parameters:
    train_acc (float): Training accuracy
    val_acc (float): Validation accuracy

    Returns:
    bool: True if overfitting (difference > 0.10), else False
    """
    
    return (train_acc - val_acc) > 0.10


# Example call
print(is_overfitting(0.95, 0.80))  # True
print(is_overfitting(0.90, 0.85))  # False
```
